```python
import string
string.punctuation
```

```python
# now we will remove the punctuation and stopwords both
```

```python
def text_process(trip):      # creating a function
    ''' 
    1. remove the punctuation  ##  a docstring
    2. remove the stopwords 
    3. return the list of clean textwords
    
    '''
    nopunc = [ char for char in trip if char not in string.punctuation]
    nopunc = "".join(nopunc)

    return [word for word in nopunc.split() if word not in stopwords.words('english')]
```

```python
trip.Review
```

```python
trip.Review = trip.Review.str.lower()
trip.Review

# ppython is case sensitive to conver full doc either into lower case or
# upper case
```

```python
trip['Review'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#  it is used to get the count of each and every word
```

```python
bow_transformer = CountVectorizer(analyzer= text_process).fit(trip['Review'])
```

```python
bow_transformer.vocabulary_
```

```python
len(bow_transformer.vocabulary_)
```

```python
# It means that the TDm will be having 9431 cols
```

```python
trip_bow = bow_transformer.transform(trip.Review)
# over there TDm is created
trip_bow.shape
```

```python
type(trip_bow)
```

```python
from sklearn.model_selection import train_test_split

df_x_train,df_x_test, df_y_train,df_y_test = train_test_split(trip_bow, trip.Rating , test_size = 0.2 ,random_state = 123)
```

```python
print(df_x_train.shape)
print(df_y_train.shape)
print('**********************')
print(df_x_test.shape)
print(df_y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(df_x_train, df_y_train)
```

```python
pred = nb.predict(df_x_test)
```

```python
from sklearn.metrics import confusion_matrix

tab = confusion_matrix(df_y_test , pred)
tab
```
