### Sampling

```python
ppt_x = ppt.iloc[: , 0:-1]
ppt_y = ppt.iloc[: , -1 ]
```

```python
from sklearn.model_selection import train_test_split
ppt_x_train,ppt_x_test, ppt_y_train,ppt_y_test = train_test_split(ppt_x,ppt_y ,test_size = 0.2, random_state = 420)
```

### For Tensorflow  (NN) your data has to be in array format and not in DF format because for doing mathematical calculations we can't do on DataFrame

```python
ppt_x_train = np.array(ppt_x_train)
ppt_x_test  = np.array(ppt_x_test)
```

```python
ppt_y_train = np.array(ppt_y_train)
ppt_y_test  = np.array(ppt_y_test)
```

```python
ppt_x_train.shape
```

```python
def build_model():
    model = tf.keras.Sequential([                                                  
         tf.keras.layers.Dense(64, activation = tf.nn.relu , input_dim = 74),  # 1st hidden layer
         tf.keras.layers.Dense(64, activation = tf.nn.relu), # 2nd hidden layer
         tf.keras.layers.Dense(64, activation = tf.nn.relu), # 3rd hidden layer
         tf.keras.layers.Dense(64, activation = tf.nn.relu), # 4th hidden layer
         tf.keras.layers.Dense(1) # o/p layer (numeric and continuous dont use softmax and due to n and c neuron = 1)
    ])
    
    optimizer = tf.keras.optimizers.RMSprop(0.005)
    
    model.compile(loss = 'mse',
                  optimizer = optimizer,
                  metrics = ['mae' , 'mse'])
    return model
    
```

```python
model = build_model()
model.fit(ppt_x_train , ppt_y_train, epochs = 1000)
```
