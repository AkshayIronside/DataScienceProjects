```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
scaled_snsdata = scalar.fit_transform(snsdata)
```

```python
## scaled_snsdata
```

```python
from sklearn.preprocessing import normalize
```

```python
scaled_snsdata = normalize(scaled_snsdata)
```

```python
dbscan = DBSCAN(eps=0.8,min_samples=10)
```

```python
model_db = dbscan.fit(scaled_snsdata)
```

```python
# how many clusters have been created???
# inn DBSCAN number of clusters are not direcly controlled
# but they are controlled by min distance and min number of points
```
