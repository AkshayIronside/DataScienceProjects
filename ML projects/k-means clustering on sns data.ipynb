```python
# build the plot like in knn
# find the value of ssd from k = 1 till 10
```

```python
ssd = []
for k in range(1 , 11):
    kmeans_sns = KMeans(n_clusters = k)
    kmeans_sns.fit(snsdata)
    scr = kmeans_sns.score(snsdata)
    scr = np.round(abs(scr))
    ssd.append(scr)
```

```python
ssd
```

```python
numbers = list(range(1,11))
```

```python
plt.figure(figsize = (6,6) )
plt.plot(numbers , ssd , marker = '*',color = 'r')
plt.grid()
```
